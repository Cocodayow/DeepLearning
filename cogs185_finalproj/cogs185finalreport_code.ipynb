```python
import requests
import string
import random
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import torch.optim as optim

# URL for the complete works of Shakespeare from Project Gutenberg
url = 'https://www.gutenberg.org/files/100/100-0.txt'

# Send a GET request to download the text
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    file_content = response.text
    # Save the content to a file
    with open('complete_shakespeare.txt', 'w', encoding='utf-8') as f:
        f.write(file_content)
    print("File downloaded and saved as 'complete_shakespeare.txt'")
else:
    print("Failed to retrieve the file. Status code:", response.status_code)

# Read the complete works
with open('complete_shakespeare.txt', 'r', encoding='utf-8') as f:
    file_content = f.read()

# Filter out non-printable characters
filtered_content = ''.join(filter(lambda x: x in string.printable, file_content))

# Optionally, save the filtered content back to a file
with open('filtered_shakespeare.txt', 'w', encoding='utf-8') as f:
    f.write(filtered_content)

print("Filtered file created with only printable characters.")

# Prepare for Dataset
all_chars = string.printable
n_chars = len(all_chars)
file = open('./filtered_shakespeare.txt').read()
file_len = len(file)

print('Length of file: {}'.format(file_len))
print('All possible characters: {}'.format(all_chars))
print('Number of all possible characters: {}'.format(n_chars))

# Helper functions
def random_sequence(seq_len=50):
    start = random.randint(0, file_len - seq_len)
    end = start + seq_len + 1
    return file_content[start:end]

def sequence_to_onehot(seq):
    tensor = torch.zeros(len(seq), n_chars)
    for i, char in enumerate(seq):
        index = all_chars.find(char)
        if index != -1:
            tensor[i][index] = 1
    return tensor.unsqueeze(0)

def get_batch(batch_size):
    seq = random_sequence()
    input_seq = seq[:-1]
    target_seq = seq[1:]
    input_tensor = sequence_to_onehot(input_seq)
    target_tensor = sequence_to_index(target_seq)
    input_tensor = input_tensor.repeat(batch_size, 1, 1)
    target_tensor = target_tensor.repeat(batch_size).view(-1)
    return input_tensor, target_tensor

def sequence_to_index(seq):
    tensor = torch.zeros(len(seq), dtype=torch.long)
    for i, char in enumerate(seq):
        index = all_chars.find(char)
        tensor[i] = index if index != -1 else -1
    return tensor

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)

class GRUModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, activation='relu'):
        super(GRUModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.rnn = nn.GRU(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x, hidden):
        out, hidden = self.rnn(x, hidden)
        out = self.fc(out)
        return out, hidden

    def init_hidden(self, batch_size):
        return torch.zeros(1, batch_size, self.hidden_dim).to(device)

class LSTMModel(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, activation='relu'):
        super(LSTMModel, self).__init__()
        self.hidden_dim = hidden_dim
        self.rnn = nn.LSTM(input_dim, hidden_dim, batch_first=True)
        self.fc = nn.Linear(hidden_dim, output_dim)

    def forward(self, x, hidden):
        out, hidden = self.rnn(x, hidden)
        out = self.fc(out)
        return out, hidden

    def init_hidden(self, batch_size):
        return (torch.zeros(1, batch_size, self.hidden_dim).to(device),
                torch.zeros(1, batch_size, self.hidden_dim).to(device))

# Wrapper classes
class GRUWrapper(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, activation='relu'):
        super(GRUWrapper, self).__init__()
        self.model = GRUModel(input_dim, hidden_dim, output_dim, activation)

    def forward(self, x, hidden):
        return self.model(x, hidden)

    def init_hidden(self, batch_size):
        return self.model.init_hidden(batch_size)

class LSTMWrapper(nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim, activation='relu'):
        super(LSTMWrapper, self).__init__()
        self.model = LSTMModel(input_dim, hidden_dim, output_dim, activation)

    def forward(self, x, hidden):
        return self.model(x, hidden)

    def init_hidden(self, batch_size):
        return self.model.init_hidden(batch_size)

#adam: 0.003
#sgd: 0.01 momentum:0.9
def create_optimizer(optimizer_choice, model):
    if optimizer_choice == 'adam':
        return optim.Adam(model.parameters(), lr=0.003)
    elif optimizer_choice == 'sgd':
        return optim.SGD(model.parameters(), lr=0.01, momentum=0.9)
    elif optimizer_choice == 'rmsprop':
        return optim.RMSprop(model.parameters(), lr=0.001)


def train(model, optimizer, criterion, input, target):
    model.zero_grad()
    hidden = model.init_hidden(input.size(0))
    output, hidden = model(input, hidden)
    output = output.view(-1, n_chars)
    loss = criterion(output, target)
    loss.backward()
    optimizer.step()
    return loss.item()

def evaluate(model, seed='W', length=100):
    batch_size = 1
    hidden = model.init_hidden(batch_size)
    input_seq = sequence_to_onehot(seed).to(device)
    generated_seq = seed
    for _ in range(len(seed) - 1):
        output, hidden = model(input_seq[:, _, :], hidden)
    input_seq = input_seq[:, -1:, :]
    for _ in range(length):
        output, hidden = model(input_seq, hidden)
        prob_dist = torch.softmax(output[:, -1, :], dim=-1)
        predicted_index = torch.multinomial(prob_dist, 1)[0]
        predicted_char = all_chars[predicted_index]
        generated_seq += predicted_char
        input_seq = sequence_to_onehot(predicted_char).to(device)
    return generated_seq

# Run experiments
def run_experiment(model_type, optimizer_choice, iterations=50000, print_every=1000):
    if model_type == 'GRU':
        model = GRUWrapper(input_dim=n_chars, hidden_dim=128, output_dim=n_chars, activation='relu').to(device)
    elif model_type == 'LSTM':
        model = LSTMWrapper(input_dim=n_chars, hidden_dim=128, output_dim=n_chars, activation='relu').to(device)

    optimizer = create_optimizer(optimizer_choice, model)
    criterion = nn.CrossEntropyLoss(ignore_index=-1)

    print(f"Running {model_type} model with {optimizer_choice} optimizer")

    losses = []
    total_loss = 0

    for i in range(iterations):
        input_batch, target_batch = get_batch(batch_size)
        input_batch, target_batch = input_batch.to(device), target_batch.to(device)
        loss = train(model, optimizer, criterion, input_batch, target_batch)
        total_loss += loss

        if (i + 1) % print_every == 0:
            avg_loss = total_loss / print_every
            losses.append(avg_loss)
            total_loss = 0
            print(f'Iteration {i+1}/{iterations}, Loss: {avg_loss}')
            print('Generated text:', evaluate(model, length=200))

    # Plot training loss
    plt.plot(losses)
    plt.xlabel('Iterations (x1000)')
    plt.ylabel('Loss')
    plt.title(f'Training Loss for {model_type} with {optimizer_choice}')
    plt.show()

# Hyperparameters
batch_size = 32
hidden_dim = 128
iterations = 50000
print_every = 1000

# Run experiments for GRU and LSTM with different optimizers
for model_type in ['LSTM', 'GRU']:
    for optimizer_choice in ['sgd', 'adam']:
        run_experiment(model_type, optimizer_choice, iterations, print_every)
```
